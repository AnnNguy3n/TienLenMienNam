```python
import gym
import gym_TLMN
from colorama import Fore, Style


def main():
    env = gym.make('gym_TLMN-v0')
    env.reset()

    # print([i.name for i in env.players])

    for i in range(500):
        env.render()

        o,a,done,t = env.step(env.turn.action(env.dict_input))
        if done:
            break

    print(Fore.LIGHTGREEN_EX, 'Win, name:', Fore.LIGHTMAGENTA_EX, env.p_name_victory, Fore.LIGHTCYAN_EX, '0', end='   ')
    for p in env.players:
        if p.name != env.p_name_victory:
            print(Fore.LIGHTRED_EX, 'Lose, name:', Fore.LIGHTMAGENTA_EX, p.name, Fore.LIGHTCYAN_EX, p.amount_cards_remaining, end='   ')
    
    print(Style.RESET_ALL)

    for i in range(env.players.__len__()):
        env.render()

        o,a,done,t = env.step(env.turn.action(env.dict_input))

    return env.p_name_victory
```

```python
main()
```
